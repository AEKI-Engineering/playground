```python
class DataLoader():
    def load_to_pandas_df(self, dataset_name):
        datasets_folder = os.getcwd() + "\\" + "test_data"
        subfolders = [ f.path for f in os.scandir(datasets_folder) if f.is_dir()]
        available_datasets = [dataset_subfolder.split('\\')[-1] for dataset_subfolder in subfolders]
        if(not dataset_name in available_datasets):
            print("DATASET UNKNOWN - PLEASE REFER TO ONE OF THE GIVEN DATASETS: ")
            for available_dataset in available_datasets:
                print(available_dataset)
            return None
        COLUMNS = ["path", "class", "pixel_array", "dimensions"]
        return_dataframe = pd.DataFrame(columns=COLUMNS)
        selected_dataset = datasets_folder + "\\" + dataset_name
        class_folders = [ f.path for f in os.scandir(selected_dataset) if f.is_dir()]
        for class_folder in class_folders:
            files = [ f.path for f in os.scandir(class_folder) if f.is_file() ]
            object_class = class_folder.split('\\')[-1]
            for file in files:
                image = cv2.imread(file)
                image_array = np.asarray(image)
                dimensions = image.shape
                return_dataframe = return_dataframe.append({"path": file, "class": object_class, "pixel_array": image_array, "dimensions": dimensions}, ignore_index=True)
            
        return return_dataframe

                
            
        
        

```

```python
class MyFirstNeuralNetowrk(nn.Module):
    def __init__(self):
        self.flatten = nn.Flatten()
        self.linear_relu_stack = nn.Sequential(
            nn.Linear(28*28, 512),
            nn.ReLU(),
            nn.Linear(512, 512),
            nn.ReLU(),
            nn.Linear(512, 10),
            nn.ReLU()
        )

    def forward(self, x):
        x = self.flatten(x)
        logits = self.linear_relu_stack(x)
        return logits
```
